# Modeling Process

**Cross-industry standard process for data mining**, known as **CRISP-DM**, is an open standard process model that describes common approaches used by data mining experts. It is the most widely-used analytics model.

6 high level phases of the data mining process:
- Business Understanding
- Data Understanding
- Data Preparation
- Modeling
- Evaluation
- Deployment

The sequence of the phases is not strict and moving back and forth between different phases as it is always required.

## Business Understanding
For our business problem, we only want to look at movies that were released in US theaters. So, let's subset our data to only include relevant movies. We  We don't have a data column that tells us this exactly, so we will try our best to aproximate this.  

First, let's limit our dataset to only movies created in the US.

```python
df['country'].value_counts()
```

```python
#your code here

df= df[df['country']=='USA']
```

```python
df.shape
```

Now let's only use movies that have been given a MPAA movie rating. This should help remove all 'made for TV' movies. 

```python
df['content_rating'].value_counts()
# a lot of ratings for non-movie thearters
# so we need to remove them
```

```python
df = df[df['content_rating'].isin(['R', 'PG-13', 'PG', 'G'])]
```

Only looking at movies that are rated R, PG-13, PG, and G.

```python
df.shape
```

Also, we want to remove independent movies that were only released in a few theaters. Again, we don't have data on this, so we will rmeove movies that made less than $1 million dollars. 

```python
df = df[df['gross']>1000000]
```

Only use movies that made over $1 Million.

Finally, we can't predict the gross of the movie if we are missing that data. So we will drop all observatiosn missing that data. 

```python
df.shape
```

```python
df.dropna(subset=['gross'], inplace=True) # drop ones that are missing a gross
```

```python
df.shape
```

## Modeling 


Now that our data is cleaned up and prepared, we cna start the modeling process. First step is to do a train-test split so that we can evaluate the model after fitting it.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

Split ploymnomial test data and target data into x_train and y_train, respectivly.

```python
#call train_test_split on the data and capture the results
X_train, X_test, y_train, y_test = train_test_split(df_poly, df['gross'], random_state=34,test_size=0.2)

#check the shape of the results
print("Training set - Features: ", X_train.shape, "Target: ", y_train.shape)
print("Training set - Features: ", X_test.shape, "Target: ",y_test.shape)
```

I'm going to use a LASSO model later so I want to scale my data. Remember to fit your scaler to the training data and then transform the test data.

**Scale the data so you can use LASSO later.**

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()


#fit and transform the training data
X_train = pd.DataFrame(data=scaler.fit_transform(X_train), columns=df_poly.columns)

#transform the testing dat
X_test = pd.DataFrame(data=scaler.transform(X_test), columns=df_poly.columns)
```

```python
# instantiate a linear regression object
lm_2 = LinearRegression()

#fit the linear regression to the data
lm_2 = lm_2.fit(X_train, y_train)

y_train_pred = lm_2.predict(X_train)

train_rmse = np.sqrt(metrics.mean_squared_error(y_train, y_train_pred))


print('Training Root Mean Squared Error:' , train_rmse)
```

*Our RMSE is now $52 mil.*

Check residuals to see if they are normally distributed. 

```python
residuals = (y_train- y_train_pred)

plt.hist(residuals, bins=30)
```

There is a high tail of big misses. We can look into it by taking the log or something else...

Is there a relationship?

```python
sns.residplot(y_train_pred, y_train, lowess=True, color="g")
```

The relationship is not that bad. We can keep tinkering with it too.

### Feature Selection

**VIF**
Run VIF on data *before* doing the polynomials.

Or run VIF on **df[features]**.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor


[variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]

vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X_train.values, i) for i in range(X_train.shape[1])]
vif["features"] = X_train.columns
vif.round(1)
```

Run VIF on df[features] to get a better view. (we don't need to see the VIF of all of our polynomials)

```python
[variance_inflation_factor(df[features].values, i) for i in range(df[features].shape[1])]

vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(df[features].values, i) for i in range(df[features].shape[1])]
vif["features"] = df[features].columns
vif.round(1)
```

VIF is the value of a feature than can be explained by another feature. The vif for *actor_1_facebook_likes* is probably very high due to *cast_total_facebook_likes*. So if the latter were to be removed, the vif for the former would go down becuase it can be explained by one less feature. 
 

Check VIF: If there is a very high vif that I didn't expect, this would be a quick way to find and remove them ahead of time.

```python
vif.sort_values(by = 'VIF Factor', ascending = False)
```

Quicker way to sort the vif factor. Could also do a seletkbest.

Throw it into recursive feature elimination.

```python
from sklearn.feature_selection import RFECV

ols = LinearRegression()

# Create recursive feature eliminator that scores features by mean squared errors
selector = RFECV(estimator=ols, step=1, cv=7, scoring='neg_mean_squared_error', n_jobs=-1, verbose=1)

# Fit recursive feature eliminator 
selector.fit(X_train, y_train)
```

We went from 104 features down to 20.

```python
selected_rfe = X_train.columns[selector.support_]
removed_rfe = X_train.columns[~selector.support_]
```

```python
list(removed_rfe)

len(list(selected_rfe))
```

We had a lot of feautres **removed**. 19 were left and considered "acceptable".

We can look at those features.

```python
selected_rfe
```

Now let's use those selected features to fit a model.

**Fit a model to those selected feautres**

```python
#instantiate a linear regression object
lm_rfe = LinearRegression()

#fit the linear regression to the data
lm_rfe = lm_rfe.fit(X_train[selected_rfe], y_train)

y_rfe = lm_rfe.predict(X_train[selected_rfe])


trainRFE_rmse = np.sqrt(metrics.mean_squared_error(y_train, y_rfe))


print('Training Root Mean Squared Error:' , trainRFE_rmse)

y_pred_rfe = lm_rfe.predict(X_test[selected_rfe])

testRFE_rmse = np.sqrt(metrics.mean_squared_error(y_test, y_pred_rfe))

print('Testing Root Mean Squared Error:' , testRFE_rmse)
```

Testing set MSE is way down to $48 mil. Therefore we are confident that we are **not** overfitting the data to the training set. 

**Try with a LASSO model**

```python
## training the model
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=.00001, normalize=False)

lasso.fit(X_train,y_train)

y_train_pred_lasso = lasso.predict(X_train)
y_pred_lasso = lasso.predict(X_test)

train_rmse_lasso01 = metrics.mean_absolute_error(y_train, y_train_pred_lasso)
test_rmse_lasso01 = np.sqrt(metrics.mean_squared_error(y_test, y_pred_lasso))
print('Training Error: '+ str(train_rmse_lasso01) )
print('Testing Error: '+ str(test_rmse_lasso01) )

```

LASSO does horribly with a test RMSE  of $90 mil. 

This is happening because we gave it the entire poly df. In that poly df, called x_train, we have a lot of features that are highly coorelated. AKA columns that have many values of zero or no impact on our data or are exact dupliactes of each other. 

This breaks LASSO because it doesn't know which column to associate it with. 

So to try to remedy this, we can lower the amount of features we give to LASSO. Down-select to our top 50 columns using **selectkBest**

```python
from sklearn.feature_selection import SelectKBest, f_regression,mutual_info_regression

selector = SelectKBest(f_regression, k=30)

selector.fit(X_train, y_train)
selected_columns = X_train.columns[selector.get_support()]
removed_columns = X_train.columns[~selector.get_support()]
```

```python
## training the model
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=0.1, normalize=False)

lasso.fit(X_train[selected_columns],y_train)

y_train_pred = lasso.predict(X_train[selected_columns])
y_pred_lasso = lasso.predict(X_test[selected_columns])

train_rmse_lasso01 = metrics.mean_absolute_error(y_train, y_train_pred)
test_rmse_lasso01 = np.sqrt(metrics.mean_squared_error(y_test, y_pred_lasso))
print('Training Error: '+ str(train_rmse_lasso01) )
print('Testing Error: '+ str(test_rmse_lasso01) )


```

Our test RMSE is down to $64 mil, so that's a little better.

You can play around with the top number of features and/or the alpha value.
