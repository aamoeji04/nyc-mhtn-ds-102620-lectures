```python
master_movie_bom = []
for movie in bom_tr[1:]:
    movie_info = []
    names = movie.find('a').get_text()
    t_gross = movie.findAll(class_ = 'a-text-right')
    t_gross_vals = [movie.text for movie in t_gross[4:6]]
    movie_info.append(names)
    movie_info.extend(t_gross_vals)
    master_movie_bom.append(movie_info)
print(master_movie_bom)
```
